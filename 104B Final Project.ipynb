### A basic neural network

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#neural net
def train(x, y, weights, alpha, epochs):
    m = x.shape[1]
    costs = []
    indexes = []
    for i in range(epochs):
        indexes.append(i)
        #prediction
        y_hat = np.dot(x, weights)
        #error
        loss = y_hat - y
        cost = np.sum(loss ** 2) / (2*m)
        costs.append(cost)
        #gradient descent
        gradient = np.dot(x.transpose(), loss) / m
        # update
        weights = weights - alpha * gradient
        
        print("Iteration: {} | Cost: {}".format(i, cost))
        
    return  y_hat, costs, indexes 
  
x = np.linspace(0, 6*np.pi, 1000)
x_shaped = x.reshape(1, 1000)
y = np.sin(x) + 0.1 * np.random.randn(x_shaped.shape[1])

weights = np.random.randn(x_shaped.shape[1], 1) * 0.1
learning_rate = 0.0001
epochs = 500

y_hat, costs, indexes = train(x_shaped, y, weights, learning_rate, epochs)
```

```python
plt.figure()
plt.scatter(x, y)
plt.title("Generated Sine Wave Data")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
plt.figure()
plt.scatter(x, y, label="Target")
plt.plot(x, y_hat.transpose(), c='red', label="Predicted")
plt.legend()
plt.title("Fitted Data with Generated Data")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
plt.figure()
plt.plot(indexes, costs)
plt.title("Epoch vs. Cost")
plt.xlabel("Epoch")
plt.ylabel("Cost")
plt.show()
```
